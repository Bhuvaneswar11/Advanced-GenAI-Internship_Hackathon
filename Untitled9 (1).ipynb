```python
import pandas as pd
orders_df=pd.read_csv("orders.csv")
print(orders_df.head())
```

```python
users_df=pd.read_json("users.json")
print(users_df.head())
```

```python
orders_users_df=pd.merge(
    orders_df,
    users_df,
    on="user_id",
    how="left"
)
```

```python
import sqlite3
import pandas as pd

conn=sqlite3.connect("restaurants.db")
conn.execute("Drop Table if exists restaurants;")

with open("restaurants.sql","r") as f:
    sql_script=f.read()

    conn.executescript(sql_script)

    restaurants_df=pd.read_sql("select* from restaurants",conn)
    restaurants_df.head()

```

```python
final_df=pd.merge(
    orders_users_df,
    restaurants_df,
    on="restaurant_id",
    how="left"

)
```

```python
final_df.to_csv("final_food_delivery_dataset.csv",index=False)
print("Final dataset created successfully!")
```

```python
final_df[final_df['membership']=='Gold']\
.groupby('city')['total_amount']\
.sum()\
.sort_values(ascending=False)
```

```python
final_df.groupby('cuisine')['total_amount']\
.mean()\
.sort_values(ascending=False)
```

```python
user_spend=final_df.groupby('user_id')['total_amount'].sum()
user_spend[user_spend > 1000].count()
```

```python
final_df['rating_range']=pd.cut(
    final_df['rating'],
    bins=[3.0,3.5,4.0,4.5,5.0],
    labels=['3.0-3.5','3.6-4.0','4.1-4.5','4.6-5.0']
)

final_df.groupby('rating_range')['total_amount']\
.sum()\
.sort_values(ascending=False)
```

```python
final_df[final_df['membership']=='Gold']\
.groupby('city')['total_amount']\
.mean()\
.sort_values(ascending=False)
```

```python
cuisine_stats=final_df.groupby('cuisine').agg(
    restaurants=('restaurant_name_y','nunique'),
    revenue=('total_amount','sum')
)
cuisine_stats.sort_values(['restaurants','revenue'],ascending=[True,False])
```

```python
gold_orders=final_df[final_df['membership']=='Gold']['order_id'].count()
total_orders=final_df['order_id'].count()
round((gold_orders/total_orders)*100)
```

```python
restaurant_stats = final_df.groupby('restaurant_name_y').agg(
    avg_order=('total_amount', 'mean'),
    total_orders=('order_id', 'count')
)

restaurant_stats[restaurant_stats['total_orders'] < 20] \
    .sort_values('avg_order', ascending=False)
```

```python
final_df['order_date'] = pd.to_datetime(final_df['order_date'])
final_df['quarter'] = final_df['order_date'].dt.to_period('Q')

final_df.groupby('quarter')['total_amount'] \
    .sum() \
    .sort_values(ascending=False)
```
